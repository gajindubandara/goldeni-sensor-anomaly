```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error
```

```python
# Read data
train_data = pd.read_csv("train.csv")
```

```python
# EDA - Exploratory Data Analysis
print("First few rows of the dataset:")
print(train_data.head())
```

```python
# Summary statistics
print("Summary statistics:")
print(train_data.describe())
```

```python
# Checking data types
print("Data types of columns:")
print(train_data.dtypes)
```

```python
# Checking for missing values
print("Missing values:")
print(train_data.isnull().sum())
```

```python
# Data visualization
# Distribution of numerical features
train_data.hist(figsize=(12, 8))
plt.suptitle("Histogram of Numerical Features")
plt.show()
```

```python
# Box plot for numerical features
plt.figure(figsize=(12, 8))
sns.boxplot(data=train_data.drop(['PassengerId'], axis=1))
plt.title("Box Plot of Numerical Features")
plt.show()
```

```python
# Count plot for categorical features
plt.figure(figsize=(12, 8))
sns.countplot(data=train_data, x='Survived')
plt.title("Count Plot of Survived")
plt.show()
```

```python
# Data preprocessing
train_data.replace({'Sex': {'male': 0, 'female': 1}, 'Embarked': {'S': 0, 'C': 1, 'Q': 2}}, inplace=True)
train_data['Age'].fillna(train_data['Age'].mean(), inplace=True)
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace=True)

# Splitting data into features and target
y = train_data.Survived.copy()
train_data = train_data.drop(['Name', 'Ticket', 'Survived', 'Cabin'], axis=1)
```

```python
# Correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(train_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
# Splitting train set
X_train, X_test, y_train, y_test = train_test_split(train_data, y, test_size=0.2, random_state=123)
```

```python
# Define and initialize the classifiers
decision_tree_classifier = DecisionTreeClassifier(random_state=123)
random_forest_classifier = RandomForestClassifier(random_state=123)
logistic_regression_classifier = LogisticRegression(random_state=123)

#  Train the classifiers (Decision tree)
decision_tree_classifier.fit(X_train, y_train)
```

```python
#  Train the classifiers (random forest)
random_forest_classifier.fit(X_train, y_train)
```

```python
# # Train the classifiers (logistic regression)
logistic_regression_classifier.fit(X_train, y_train)
```

```python
# Function to evaluate the model
def evaluate_model(model, X_test, y_test):
    """ Evaluate the performance of the model on the test set."""
    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")

    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    print("\nConfusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

    return accuracy
```

```python
# Evaluate the classifiers
print("Decision Tree Classifier")
decision_tree_accuracy  = evaluate_model(decision_tree_classifier, X_test, y_test)

print("Random Forest Classifier")
random_forest_accuracy = evaluate_model(random_forest_classifier, X_test, y_test)

print("Logistic Regression Classifier")
logistic_regression_accuracy = evaluate_model(logistic_regression_classifier, X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV

# Define hyperparameters for Decision Tree Classifier
dt_param_grid = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Define hyperparameters for Random Forest Classifier
rf_param_grid = {
    'n_estimators': [10, 50, 100, 200],
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2'],
    'bootstrap': [True, False]
}

# Define hyperparameters for Logistic Regression Classifier
lr_param_grid = {
    'penalty': ['l1', 'l2'],
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'fit_intercept': [True, False],
    'solver': ['liblinear', 'lbfgs', 'newton-cg', 'sag', 'saga']
}
```

```python
# Perform hyperparameter tuning using GridSearchCV
dt_grid_search = GridSearchCV(DecisionTreeClassifier(random_state=123), dt_param_grid, cv=5, scoring='accuracy', verbose=1)
dt_grid_search.fit(X_train, y_train)
```

```python
rf_grid_search = GridSearchCV(RandomForestClassifier(random_state=123), rf_param_grid, cv=5, scoring='accuracy', verbose=1)
rf_grid_search.fit(X_train, y_train)
```

```python
lr_grid_search = GridSearchCV(LogisticRegression(random_state=123), lr_param_grid, cv=5, scoring='accuracy', verbose=1)
lr_grid_search.fit(X_train, y_train)
```

```python
# Get the best models from GridSearchCV
hy_dt_model = dt_grid_search.best_estimator_
hy_rf_model = rf_grid_search.best_estimator_
hy_lr_model = lr_grid_search.best_estimator_

print("Decision Tree Classifier")
hy_dt_model_accurancy = evaluate_model(hy_dt_model, X_test, y_test)
print("Random Forest Classifier")
hy_rf_model_accurancy = evaluate_model(hy_rf_model, X_test, y_test)
print("Logistic Regression Classifier")
hy_lr_model_accurancy = evaluate_model(hy_lr_model, X_test, y_test)
```

```python
# Identify the best performing model
best_accuracy = max(hy_dt_model_accurancy, hy_rf_model_accurancy, hy_lr_model_accurancy)

if best_accuracy == hy_dt_model_accurancy:
    best_model = hy_dt_model
    best_model_name = "Decision Tree Classifier"
elif best_accuracy == hy_rf_model_accurancy:
    best_model = hy_rf_model
    best_model_name = "Random Forest Classifier"
else:
    best_model = hy_lr_model
    best_model_name = "Logistic Regression Classifier"

print(f"Best Performing Model: {best_model_name}")
print(f"Best Accuracy: {best_accuracy:.4f}")
```

```python
#Calculate mean squared error
val_predictions = best_model.predict(X_test)
mse = mean_squared_error(y_test, val_predictions)
print(f"Mean Squared Error (MSE): {mse:.4f}")
```

```python
# Save the best model
import pickle
pickle.dump(best_model, open('./best_model.h5', 'wb'))
```

```python
# Test set predictions
test_data = pd.read_csv("test.csv")
test_passenger_ids = test_data["PassengerId"]

test_data = test_data.drop(['Name', 'Ticket', 'Cabin'], axis=1)
test_data.replace({'Sex': {'male': 0, 'female': 1}, 'Embarked': {'S': 0, 'C': 1, 'Q': 2}}, inplace=True)
test_data['Age'].fillna(test_data['Age'].mean(), inplace=True)
test_data['Fare'].fillna(test_data['Fare'].mean(), inplace=True)
test_data['Embarked'].fillna(test_data['Embarked'].mode()[0], inplace=True)

# Make predictions on the test set
test_predictions = best_model.predict(test_data)

# Create submission DataFrame
submission_df = pd.DataFrame({
    "PassengerId": test_passenger_ids,
    "Survived": test_predictions
})

# Save submission file
submission_df.to_csv("submission.csv", index=False)
print("Submission file saved successfully.")
```
